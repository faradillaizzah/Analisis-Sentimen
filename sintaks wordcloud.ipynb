```python
import pandas as pd
data = "/content/drive/MyDrive/Skripsi/Data/datawordcloud.xlsx"
df = pd.read_excel(data)
df
```

```python
# Sentimen Positif
df_pos=df[df['sentimen']=='positif']
all_words_lem = ' '.join([word for word in df_pos['fix']])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(background_color='white', width=800, height=500, random_state=1, max_font_size=130).generate(all_words_lem)

plt.figure(figsize=(20, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off');
```

```python
# Sentimen Negatif
df_neg=df[df['sentimen']=='negatif']
all_words_lem = ' '.join([word for word in df_neg['fix']])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(background_color='white', width=800, height=500, random_state=1, max_font_size=130).generate(all_words_lem)

plt.figure(figsize=(20, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off');
```
