```python
emoji = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)

def preprocessing(text):
    # Menghapus username
    text = re.sub('@[^\s]+', '', str(text))
    # Menghapus URL
    text = re.sub(r'http\S+', '',str(text))
    # Menghapus angka dan tanda baca
    text = re.sub('[0-9%s]' % re.escape(string.punctuation), '',str(text))
    # Mengubah menjadi lowercase
    text = text.lower()
    # Menghapus emoji
    text = emoji.sub(r'', str(text))
    # Menghapus words with 2 or fewer letters
    text = re.sub(r'\b\w{1,2}\b', '', text)
    # Remove whitespace (including new line characters)
    text = re.sub(r'\s\s+', ' ', text)
    # Remove characters beyond Basic Multilingual Plane (BMP) of Unicode:
    text = ''.join(c for c in text if c <= '\uFFFF')
    return text
```

```python
# clean dataframe's text column
df['cleansed']=df['full_text'].apply(preprocessing)
# preview some cleaned tweets
df
```

```python
# TOKENIZING
from nltk.tokenize import word_tokenize
nltk.download('punkt')
def tokenizingText(text): # Tokenizing or splitting a string, text into a list of tokens
 text = word_tokenize(text)
 return text
# tokenize dataframe's text column
df['tokenized'] = df['cleansed'].apply(tokenizingText)
# preview some tokenized tweets
df
```

```python
# STEMMING
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
def stemmingText(text): # Reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words
 factory = StemmerFactory()
 stemmer = factory.create_stemmer()
 text = [stemmer.stem(word) for word in text]
 return text
# stemming dataframe's text column
df['stemmed'] = df['tokenized'].apply(stemmingText)
# preview some stemmed tweets
df
```

```python
# STOPWORDS REMOVAL
nltk.download('stopwords')
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from nltk.corpus import stopwords
def filteringText(text): # Remove stopwords in a text
 listStopwords = set(stopwords.words('indonesian'))
 filtered = []
 for txt in text:
   if txt not in listStopwords:
    filtered.append(txt)
    text = filtered
 return text
# remove stopwords
df['stopwords'] = df['stemmed'].apply(filteringText)
# preview some removed stopwords
df
```

```python
def toSentence(s):
 # initialize an empty string
 str1 = " "
 # return string
 str1.join(s)
 sentence = ' '.join([str(elem) for elem in s])
 return (sentence)
# List to text
df['fix'] = df['stopwords'].apply(toSentence)
# preview list to text
df
# Export data
df.to_excel('datafinallast.xlsx')
```
